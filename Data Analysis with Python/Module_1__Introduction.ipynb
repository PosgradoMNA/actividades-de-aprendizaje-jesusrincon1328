**Read**

Utilizamos la funcion **pandas.read_csv()** para leer el archivo csv indicado entre parentesis.

Colocamos la ubicación del archivo entre comillas dentro de parentesis y de esta manera pandas va a leer el archivo trayendo un dataframe.


```python
import pandas as pd
import numpy as np
```

```python
path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv"
```

Debido a que el set de datos no incluye encabvezados colocaremos el argumento **headers = None** dentro de **read_csv()** de esta manera Pandas no colocara automaticamente la primera fila como encabezado.

```python
df = pd.read_csv(path, header=None)
df
```

```python
#podemos utilizar el metodo dataframe.head(n) para visualizar solamente la parte (n) de filas del data frame, donde (n) es un entero.
#tambien podemos utilizar dataframe.tail(n) para visualizar las ultimas n filas.
```

```python
df.head(5)
```

```python
#muestra las ultimas 10 filas del dataframe "df"
df.tail(10)
```

**Agregar encabezados(atributos)**

Primero creamos la lista "headers

```python
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
print("headers\n", headers)
```

```python
df.columns = headers
df.head(5)
```

```python
#Se requiere reeemplazar los signos "?" para de esta manera remover los valores perdidos con dropna()

df1 = df.replace("?",np.nan)
df1
```

```python
df = df1.dropna(subset=["price"],axis=0)
df.head(20)
```

```python
#muestra el nombre de las columnas del dataframe

print(df.columns)
```

```python
#Para mantener los avances y modificaciones que se relizan al data set se recomienda guardar el archivo .
df.to_csv("automobile.csv", index=False)
```

**Describe**

Con "describe" podremos obtener un resumen estadistico de cada columna(tipo numerica) como por ejemplo: media, mediana, desviación estandar, etc.

dataframe.describe()

```python
df.describe()
```

```python
#si queremos tener la descripción de todas las columnas utilizamos:

df.describe(include = "all")
```

```python
#puedes seleccionar alguna columna indicando el nombre de cada una.

#utilizando con el siguiente formato:  dataframe[[' column 1 ',column 2', 'column 3']]
```

```python
#aplica el metodo ".describe()" a las columnas "Length" and "compression-ratio"
df[["length","compression-ratio"]].describe()
```
