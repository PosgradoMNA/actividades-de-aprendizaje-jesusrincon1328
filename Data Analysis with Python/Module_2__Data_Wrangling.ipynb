Data Wrangling es el proceso de convertir los datos de su formato inicial a un formato que podria ser mejor para su analisis.



```python
import pandas as pd
import matplotlib.pylab as plt
import numpy as np
```

```python
filename =  "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv"
```

```python
#creamos una lista que sera utilizada como encabezado de las columnas
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
df = pd.read_csv(filename, names = headers)
df.head() #para visualizar las primeras 5 filas
```

```python
#los signos de interrogación que aparecieron son datos faltantes que perjudicaran el analisis posterior
```

**Pasos para trabajar con datos faltantes:**

*   Identificar los datos faltantes
*   Manejar datos faltantes
*   Corregir el formato de los datos

# **Identificación y manejo de datos faltantes**

**Identificar datos faltantes**

Convertir "?" to NaN

```python
#utilizaremos la siguiente función .replace(A, B, inplace = True) 
#para reemplazar A por B

#reemplzar "?" por NaN (Not a Number)
```

```python
df.replace("?", np.nan, inplace = True)
df.head(5)
```

Hay dos metodos que podemos utilizar para identificar los datos faltantes:

*   **.isnull()**
*   **.notnull()**

La salida es un valor booleano, "True" significa que es un dato faltantes mientras que "False" no lo es.





```python
missing_data =  df.isnull()
missing_data.head(5)
```

```python
#contar datos faltantes en cada columna
```

```python
for columns in missing_data.columns.values.tolist():
    print(missing_data[columns].value_counts())
    print(columns)
    print("")
```

Como resumen de la información anterior tenemos que cada columna tiene 205 filas y 7 de las columnas tienen datos faltantes.

1.  "normalized-losses": 41 missing data
2.  "num-of-doors": 2 missing data
3.  "bore": 4 missing data
4.  "stroke" : 4 missing data
5.  "horsepower": 2 missing data
6.  "peak-rpm": 2 missing data
7.  "price": 4 missing data

**Manejo de datos faltantes**

1.  Eliminar los datos
    * Eliminar la fila completa
    * Eliminar la columna completa
2.  Reemplazar los datos
    * Reemplazar los datos por el promedio
    * Reemplazar los datos por frecuencia
    * Reemplazar los datos basado en otra funcion

```python
#Calcular el promedio de "normalized-losses"
avg_norm_loss = df["normalized-losses"].astype("float").mean(axis=0)
print("Averge of normalized losses:", avg_norm_loss)
```

```python
#Reemplazar "NaN" con el promedio de la columna "normalized-losses" 
df["normalized-losses"].replace(np.nan, avg_norm_loss, inplace = True)
df.head(5)
```

```python
#Calcular el promedio de la columna "bore"
avg_bore = df["bore"].astype("float").mean(axis=0)
print("Average of bore:", avg_bore )
```

```python
#reemplazar "NaN" con el promedio de la columna "bore"
df["bore"].replace(np.nan, avg_bore, inplace = True)
df.head(5)
```

```python
#reemplazar "NaN" en columna "stroke" con el promedio

avg_stroke = df["stroke"].astype("float").mean(axis = 0)
print("Average of stroke:", avg_stroke)
df["stroke"].replace(np.nan, avg_stroke, inplace = True)
```

```python
#reemplazar "NaN" en columna "horsepower" con el promedio

avg_horsepower = df['horsepower'].astype('float').mean(axis=0)
print("Average horsepower:", avg_horsepower)
df['horsepower'].replace(np.nan, avg_horsepower, inplace=True)
```

```python
#reemplazar "NaN" en columna "peak-rpm" con el promedio

avg_peakrpm=df['peak-rpm'].astype('float').mean(axis=0)
print("Average peak rpm:", avg_peakrpm)
df['peak-rpm'].replace(np.nan, avg_peakrpm, inplace=True)
```

```python
#para saber cuales valores estan presentes en una columna podemos utilizar el metodo "value_counts()"
```

```python
df["num-of-doors"].value_counts()
```

```python
#tambien podemos encontrar el valor mas comun de la siguiente manera:

df["num-of-doors"].value_counts().idxmax()
```

```python
#reemplazar el valor Nan de la columna "num-of-doors" por el valor mas comun.
df["num-of-doors"].replace(np.nan, "four", inplace = True)
```

```python
#finalmente eliminemos las filas que no contienen "price data"
df.dropna(subset =["price"], axis = 0, inplace = True)

#aplicamos reset-index debido a que se eliminaron 2 filas
df.reset_index(drop=True, inplace=True)
```

```python
df.head()
```

El utimo paso en la limpieza de datos es asegurar que todos los datos esten en el formato correcto (int, float, text or other).

En Pandas utilizamos:

*   **.dtype()** para revisar el tipo de dato
*   **.astype()** para cambiar el tipo de dato



```python
df.dtypes
```

```python
#convertir los datos al formato correcto.
df[["bore", "stroke"]] = df[["bore", "stroke"]].astype("float")
df[["normalized-losses"]] = df[["normalized-losses"]].astype("int")
df[["price"]] = df[["price"]].astype("float")
df[["peak-rpm"]] = df[["peak-rpm"]].astype("float")
```

```python
df.dtypes
```

**Estandarización**

Es el proceso de transformar los datos a un formato en comun que permita realizar comparaciones significativas.

```python
#las columnas "city-mpg" y "highway-mpg" se encuentran en las unidades de mpg(millas por galon)
#transformar las columnas a la unidad L/100km (L/100km = 235 / mpg)

df.head()
```

```python
df['city-L/100km'] = 235/df["city-mpg"]
df.head()
```

```python
df["highway-mpg"] = 235/df["highway-mpg"]
df.rename(columns={'highway-mpg':'highway-L/100km'}, inplace=True)
df.head()
```

**Normalización**

Normalización es el proceso de transformar los valores de diferentes variable a un mismo rango. 

```python
#escalar las columnas  "length", "width" and "height".

df["length"] = df["length"]/df["length"].max()
df['width'] = df['width']/df['width'].max()
df["height"] = df["height"]/df["height"].max()

df[["length",'width',"height"]].head()
```

**Binning**

Agrupación es un proceso de transformación variables continuas numericas en categoricas "bins" para analisis grupal.

```python
#Utilizar el metodo "cut" de Pandas para segmentar "horsepower" dentro de 3 grupos

#Convertir el dato al formato correcto
df["horsepower"]=df["horsepower"].astype(int, copy=True)
```

```python
#Visualizar la distribución

%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
plt.pyplot.hist(df["horsepower"])

#colocar las etiquetas x/y y el titulo
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

```python
bins = np.linspace(min(df["horsepower"]), max(df["horsepower"]), 4)
bins
```

```python
group_names = ['Low', 'Medium', 'High']
```

```python
df['horsepower-binned'] = pd.cut(df['horsepower'], bins, labels=group_names, include_lowest=True )
df[['horsepower','horsepower-binned']].head(20)
```

```python
df["horsepower-binned"].value_counts()
```

```python
#Mostrar la distribución de cada agrupación
pyplot.bar(group_names, df["horsepower-binned"].value_counts())

plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

Normalmente utilizamos un histograma para mostrar la distribución de los grupos

```python
#Mostrar histograma del atrubito "horsepower" con bins =3
plt.pyplot.hist(df["horsepower"], bins = 3)

# Ingresamos las etiquetas x/y y el titulo
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

**Variables Dummy**

Convierte una variable categorica en numerica a fin de etiquetar alguna categoria.

Utilizamos variables Dummy en las las variables categoricas para poder utilizarlas en analisis de regresion lineal (proximos modulos).


```python
#Utilizaremos la columna "fuel-type" que tiene dos valores: "gas" or "diesel", se le asignaran valores a estas categorias para poder ser utilizadas en otros analisis
```

```python
df.columns
```

```python
dummy_variable_1 = pd.get_dummies(df["fuel-type"])
dummy_variable_1.head()
```

```python
#cambiar el nombre de las columnas
dummy_variable_1.rename(columns={'gas':'fuel-type-gas', 'diesel':'fuel-type-diesel'}, inplace=True)
dummy_variable_1.head()
```

```python
#une el data frame "df" y "dummy_variable_1"
df = pd.concat([df, dummy_variable_1], axis=1)
df.head()
```

```python
df.drop("fuel-type", axis = 1, inplace=True)
```

```python
df.head()
```

```python
#crea una variable Dummy para la columna "aspiration"

dummy_variable_2 = pd.get_dummies(df['aspiration'])
```

```python
#cambiar el nombre de las columnas para mayor entendimiento
dummy_variable_2.rename(columns={'std':'aspiration-std', 'turbo': 'aspiration-turbo'}, inplace=True)
```

```python
dummy_variable_2.head()
```

```python
#une el data frame "df" y "dummy_variable_2"
df = pd.concat([df, dummy_variable_2], axis=1)
```

```python
df.drop('aspiration', axis = 1, inplace=True)
```
