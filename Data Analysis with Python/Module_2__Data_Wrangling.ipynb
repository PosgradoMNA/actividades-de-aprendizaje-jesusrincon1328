Data Wrangling es el proceso de convertir los datos de su formato inicial a un formato que podria ser mejor para su analisis.



```python
import pandas as pd
import matplotlib.pylab as plt
import numpy as np
```

```python
filename =  "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv"
```

```python
#creamos una lista que sera utilizada como encabezado de las columnas
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
df = pd.read_csv(filename, names = headers)
df.head() #para visualizar las primeras 5 filas
```

```python
#los signos de interrogación que aparecieron son datos faltantes que perjudicaran el analisis posterior
```

**Pasos para trabajar con datos faltantes:**

*   Identificar los datos faltantes
*   Manejar datos faltantes
*   Corregir el formato de los datos

# **Identificación y manejo de datos faltantes**

**Identificar datos faltantes**

Convertir "?" to NaN

```python
#utilizaremos la siguiente función .replace(A, B, inplace = True) 
#para reemplazar A por B

#reemplzar "?" por NaN (Not a Number)
```

```python
df.replace("?", np.nan, inplace = True)
df.head(5)
```

Hay dos metodos que podemos utilizar para identificar los datos faltantes:

*   **.isnull()**
*   **.notnull()**

La salida es un valor booleano, "True" significa que es un dato faltantes mientras que "False" no lo es.





```python
missing_data =  df.isnull()
missing_data.head(5)
```

```python
#contar datos faltantes en cada columna
```

```python
for columns in missing_data.columns.values.tolist():
    print(missing_data[columns].value_counts())
    print(columns)
    print("")
```

Como resumen de la información anterior tenemos que cada columna tiene 205 filas y 7 de las columnas tienen datos faltantes.

4.  "stroke" : 4 missing data

**Manejo de datos faltantes**

1.  Eliminar los datos
    * Eliminar la fila completa
    * Eliminar la columna completa
2.  Reemplazar los datos
    * Reemplazar los datos por el promedio
    * Reemplazar los datos por frecuencia
    * Reemplazar los datos basado en otra funcion

```python
#Calcular el promedio de "normalized-losses"
avg_norm_loss = df["normalized-losses"].astype("float").mean(axis=0)
print("Averge of normalized losses:", avg_norm_loss)
```

```python
#Reemplazar "NaN" con el promedio de la columna "normalized-losses" 
df["normalized-losses"].replace(np.nan, avg_norm_loss, inplace = True)
df.head(5)
```

```python
#Calcular el promedio de la columna "bore"
avg_bore = df["bore"].astype("float").mean(axis=0)
print("Average of bore:", avg_bore )
```

```python
#reemplazar "NaN" con el promedio de la columna "bore"
df["bore"].replace(np.nan, avg_bore, inplace = True)
df.head(5)
```

```python
#reemplazar "NaN" en columna "stroke" con el promedio

avg_stroke = df["stroke"].astype("float").mean(axis = 0)
print("Average of stroke:", avg_stroke)
df["stroke"].replace(np.nan, avg_stroke, inplace = True)
```

```python
#reemplazar "NaN" en columna "horsepower" con el promedio

avg_horsepower = df['horsepower'].astype('float').mean(axis=0)
print("Average horsepower:", avg_horsepower)
df['horsepower'].replace(np.nan, avg_horsepower, inplace=True)
```

```python
#reemplazar "NaN" en columna "peak-rpm" con el promedio

avg_peakrpm=df['peak-rpm'].astype('float').mean(axis=0)
print("Average peak rpm:", avg_peakrpm)
df['peak-rpm'].replace(np.nan, avg_peakrpm, inplace=True)
```
